```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/data/iivanova-23/data/wild_wiki.csv")
df.head()
```

```python
author_id_map = (
    df[["label", "author_name"]]
    .drop_duplicates()
    .set_index("label")
    .to_dict()["author_name"]
)

author_id_map
```

```python
sorted_author_id_map = dict(sorted(author_id_map.items()))
sorted_author_id_map
```

```python
import yaml

with open("configs/runs/author_id_map.yaml", "w") as f:
    yaml.dump(sorted_author_id_map, f)
```
